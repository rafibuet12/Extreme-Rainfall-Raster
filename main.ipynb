#Read Image (Rasterio)

```python
import rasterio
import numpy as np
```

```python
#rows, cols = rasterio.open(input_files[0]).read(1).shape #reference shape
#outputArray = [[0 for i in range(cols)] for j in range(rows)]
```

```python
count=0
dataArrayList=[]
for i in input_files:
  dataset = rasterio.open(i)
  data_array=dataset.read(1)
  #print(data_array.shape)
  dataArrayList.append(data_array)
  count=count+1
  print(count)
dataRecordsArray=np.array(dataArrayList)
```

```python
dataRecordsArray.shape
```

```python
rows, cols = (dataRecordsArray.shape[1],dataRecordsArray.shape[2]) #reference shape
outputArray=np.zeros((rows,cols))
outputArray
```

```python
#dataArrayList[4][0][1]
```

```python
#dim1 = len(dataArrayList)
#dim2 = len(dataArrayList[0])
#dim3 = len(dataArrayList[0][0])

#dim1,dim2,dim3
```

```python
#list convention
#tempList=[]
#for i in range(0, rows):
#  for j in range (0,cols):
#    for k in range(0,len(dataArrayList)):
#      tempList.append(dataArrayList[k][i][j])
#    tempList.sort()
#    value= np.percentile(tempList,90)
#    outputArray[i][j]= value
#    tempList.clear()
```

```python
tempList=[]
for i in range(0, rows):
  for j in range (0,cols):
    for k in range(0,dataRecordsArray.shape[0]):
      tempList.append(dataRecordsArray[k][i][j])
    tempList.sort()
    value= np.percentile(tempList,90)
    outputArray[i][j]= value
    tempList.clear()


```

```python
outputArray
```

#Write Output Image (Rasterio)

### Setting Destination Path

```python
outputFileName='out.tif'
output_path=os.path.dirname(os.path.dirname(path))+'/'+outputFileName
output_path
```

### Creating Rasterio Output File

```python
ref=rasterio.open(input_files[0])
ref.dtypes
#outputArray.dtype
```

```python
ref.meta
```

```python
ref.read(1).dtype
```

```python
with rasterio.open(output_path,'w',driver=ref.driver,height=ref.height,width=ref.width,count=ref.count,crs=ref.crs,transform=ref.transform,dtype=outputArray.dtype) as dest:
  dest.write(outputArray,1)
```

```python
outputFile=rasterio.open(output_path)
print(output_path)
```

```python
show(outputFile,title='Output data')
```
